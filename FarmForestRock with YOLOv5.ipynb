```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Directory where YOLOv5 saves predictions (check your output path)
pred_dir = '/content/drive/MyDrive/yolov5/runs/predict-seg/exp'  # Adjust if necessary

# List all image files in the prediction directory
predicted_images = [f for f in os.listdir(pred_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Display each image one by one
for img_name in predicted_images:
    img_path = os.path.join(pred_dir, img_name)
    img = Image.open(img_path)

    # Display the image
    plt.figure(figsize=(8, 8))  # Make the image larger
    plt.imshow(img)
    plt.axis('off')
    plt.show()


```
