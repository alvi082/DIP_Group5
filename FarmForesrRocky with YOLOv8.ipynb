```python
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the YOLOv8 instance segmentation model
model = YOLO('/content/runs/segment/yolov8-segmentation/weights/best.pt')

# Path to your test images
test_folder = '/content/drive/MyDrive/LFF-yolov8/test/images'

# Get all image paths in the test folder
image_paths = [os.path.join(test_folder, img) for img in os.listdir(test_folder) if img.endswith(('.jpg', '.png', '.jpeg'))]

# Iterate over each image
for img_path in image_paths:
    # Perform inference on the image
    results = model(img_path, imgsz=256, conf=0.4)

    # Create a larger figure to display the result
    plt.figure(figsize=(10, 10))  # Adjust the figure size as needed
    plt.imshow(results[0].plot())  # Display the segmented image
    plt.axis('off')  # Hide axis
    plt.title(os.path.basename(img_path))  # Show the image filename as title

    plt.show()  # Display the image
    plt.pause(2)  # Pause for 2 seconds before showing the next image
    plt.close()  # Close the figure to free up memory

```
